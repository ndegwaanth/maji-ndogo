```python
%%sql
-- LEFT
SELECT 
    LEFT(Source_name, 6) AS Short_Name
FROM
    Water_sources_sa_2022;
```

```python
%%sql
-- RIGHT
SELECT
    RIGHT(Source_name, 6) AS Short_Name
FROM
    Water_sources_sa_2022;
```

```python
%%sql
-- substring
SELECT
    SUBSTRING(Source_name, 1, 6) AS Short_Name
FROM
    Water_sources_sa_2022;
```

```python
%%sql
SELECT
    SUBSTRING_INDEX(Source_name, ' ', -1) AS Short_Name
FROM
    Water_sources_sa_2022;
```

```python
%%sql
SELECT
    concat(LEFT(Source_name, 3), RIGHT(Source_name, 3)) AS Short_Name
FROM
    Water_sources_sa_2022;
```

```python
%%sql
SELECT
    concat(Source_name, Source_id, Water_type) AS Combined
FROM
    Water_sources_sa_2022;
```

```python
%%sql
select
    Source_name,
    replace(Source_name, 'River', 'Lake') AS Modified_Name
from
    Water_sources_sa_2022;
```

```python
%%sql
select
    Country_name
from
    Access_to_Basic_Services
where
    Country_name like '%(%)%';
```

```python
%%sql
select
    length(Country_name) as Name_Length,
    position('(' in Country_name) as Position_of_right_Paren,
    rtrim(left(Country_name, position('(' in Country_name) - 1)) as Country_Name_Without_Paren,
    length(rtrim(left(Country_name, position('(' in Country_name) - 1))) as Length_Without_Paren
from
    Access_to_Basic_Services
where
    Country_name like '%(%)%';
```
