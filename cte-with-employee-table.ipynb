```python
%%sql
WITH salary_total AS (
    select
        Department,
        sum(Salary) AS total_salary,
        count(Department) AS total_employees
    from employee
    group by Department
)
select
    Department,
    salary_total.total_salary,
    salary_total.total_employees,
    round(salary_total.total_salary / salary_total.total_employees, 2) AS average_salary
from salary_total;
```

```python
%%sql
WITH tenure_ranks AS (
    select
        First_name,
        Start_date,
        Office,
        Department,
        RANK() OVER (PARTITION BY Office ORDER BY Start_date DESC) AS tenure_rank
    from employee
)
select
    tenure_ranks.First_name,
    tenure_ranks.Start_date,
    tenure_ranks.Office,
    tenure_ranks.Department,
    tenure_ranks.tenure_rank
from tenure_ranks
where tenure_rank IN (1, 2)
order by tenure_ranks.Office, tenure_ranks.Start_date DESC;
```

```python
%%sql
WITH main_office_in_ke AS (
    select
        Start_date,
        First_name,
        Gender,
        Country,
        Office,
        Department,
        Salary
    from employee
    where Office = 'Nairobi' and Country = 'Kenya'
),
main_office_in_sa AS (
    select
        Start_date as Start_date,
        First_name,
        Gender,
        Country,
        Office,
        Department,
        Salary
    from employee
    where Office = 'Cape Town' and Country = 'South Africa'
)
select *
from main_office_in_ke
union all
select *
from main_office_in_sa;
```
