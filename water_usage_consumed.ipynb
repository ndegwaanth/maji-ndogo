```python
%%sql
CREATE TABLE Countries (
    Country_id INT PRIMARY KEY,
    Country_name VARCHAR(50),
    Population INT
);
```

```python
%%sql
INSERT INTO Countries (Country_id, Country_name, Population) VALUES
(1, 'Country A', 1000000),
(2, 'Country B', 1500000),
(3, 'Country C', 5000000),
(4, 'Country D', 2000000);
```

```python
%%sql
CREATE TABLE Water_usage (
    Country_id INT,
    Water_consumed_m3 INT
);
```

```python
%%sql
SELECT * FROM Countries;
```

```python
%%sql
INSERT INTO Water_usage (Country_id, Water_consumed_m3) VALUES
(1, 500000),
(2, 800000),
(3, 3000000),
(5, 1000000);
```

```python
%%sql
SELECT * FROM Water_usage;
```

```python
%%sql
select
    c.Country_name,
    c.Population,
    w.Water_consumed_m3
from
    Countries c
left join
    Water_usage w
on
    c.Country_id = w.Country_id;
```

```python
%%sql
select
    m.Country_name,
    m.Population,
    p.Water_consumed_m3
from
    Countries m
right join
    Water_usage p
on
    m.Country_id = p.Country_id;
```

```python
%%sql
select
    c.Country_name,
    c.Population,
    w.Water_consumed_m3
from
    Countries c
full outer join
    Water_usage w
on
    c.Country_id = w.Country_id;
```

```python
%%sql
SELECT
    c.Country_name,
    c.Population,
    NULL AS Water_consumed_m3
FROM Countries c

UNION

SELECT
    c.Country_name,
    c.Population,
    w.Water_consumed_m3
FROM Countries c
JOIN Water_usage w
    ON c.Country_id = w.Country_id;
```

```python
%%sql
SELECT
    Country_id
FROM
    Countries
INTERSECT
SELECT
    Country_id
FROM
    Water_usage;
```

```python
%%sql
SELECT
    Country_id
FROM
    Countries
EXCEPT
SELECT
    Country_id
FROM
    Water_usage;
```
