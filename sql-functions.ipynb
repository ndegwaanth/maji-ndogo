# Aggregate numeric functions                   
* MIN()
* MAX()
* AVG()
* SUM()
* COUNT()

```python
%%sql
-- MIN
select min(Pct_managed_drinking_water_services) AS min_drinking_water_services
from access_to_basic_services;
```

```python
%%sql
-- MAX
select max(Pct_managed_drinking_water_services) AS max_drinking_water_services
from access_to_basic_services;
```

```python
%%sql
-- AVG
select avg(Est_gdp_in_billions) AS avg_gdp_in_billions
from access_to_basic_services;
```

```python
%%sql
-- SUM
select sum(Est_population_in_millions) AS total_population_in_millions
from access_to_basic_services;
```

```python
%%sql
select count(DISTINCT Region) AS total_regions,
COUNT(DISTINCT Sub_region) AS total_subregions,
COUNT(DISTINCT Country_name) AS total_countries
from access_to_basic_services;
```

# Scalar Numeric Functions
* ROUND()
* SQRT()
* LOG()


```python
%%sql
SELECT Country_name,Region,Sub_region, Round(Pct_managed_drinking_water_services, 0) AS RoundPct_managed_drinking_water_services
from access_to_basic_services
WHERE Pct_managed_drinking_water_services IS NOT NULL
ORDER BY Pct_managed_drinking_water_services ASC
LIMIT 10;
```

```python
%%sql
select Country_name,Region,Sub_region, log(Pct_managed_drinking_water_services, 10) AS LogPct_managed_drinking_water_services
from access_to_basic_services
WHERE Pct_managed_drinking_water_services IS NOT NULL
ORDER BY Pct_managed_drinking_water_services ASC
LIMIT 10;
```

```python
%%sql
select Region,
avg(Pct_managed_drinking_water_services) AS avg_pct_managed_drinking_water_services
from access_to_basic_services
WHERE Pct_managed_drinking_water_services IS NOT NULL
GROUP BY Region;
```
