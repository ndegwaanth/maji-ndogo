```python
%%sql
SELECT FirstName, LastName, Title, Salary,
       MIN(Salary) OVER(PARTITION BY Title) AS Min_Salary
FROM employees;
```

```python
%%sql
SELECT FirstName, LastName, Title, Salary,
    MAX(Salary) OVER(PARTITION BY Title) AS Max_Salary
FROM employees
order by  max_salary desc;
```

```python
%%sql
select FirstName, LastName, Title, Country,
RANK() OVER(PARTITION BY Country ORDER BY Salary DESC) AS Rank_Salary
FROM employees
order by Country, Rank_Salary;
```

```python
%%sql
select FirstName, LastName, Title, Country,
DENSE_RANK() OVER(PARTITION BY Country ORDER BY Salary DESC) AS Rank_Salary
FROM employees
order by Country, Rank_Salary;
```

```python
%%sql
select FirstName, LastName, Title, Country,
DENSE_RANK() OVER(PARTITION BY Country ORDER BY Salary DESC) AS Rank_Salary,
ROW_NUMBER() OVER(PARTITION BY Country ORDER BY Salary DESC) AS Row_Num
FROM employees
order by Country, Rank_Salary;
```

```python
%%sql
select FirstName, LastName, Title, Country,
ROW_NUMBER() OVER(PARTITION BY title ORDER BY Title DESC) AS Row_Num
FROM employees
order by Title, Row_Num;
```

```python
%%sql
select FirstName, LastName, Title, Salary,
lag(Salary, 1) over(order by Salary) as Previous_Salary
from employees
order by Salary;
```

```python
%%sql
select FirstName, LastName, Title, HireDate,Salary,
lead(Salary, 1) over(partition by Title order by HireDate) as Previous_Salary
from employees
order by Title;
```

```python
%%sql
select FirstName, LastName, Title, HireDate,Salary,
lag(Salary, 1) over(partition by Title order by HireDate) as Previous_Salary
from employees
order by Title;
```

```python
%%sql
select Title, HireDate,Salary,
first_value(FirstName) over(partition by Title order by HireDate) as First_Hired
from employees
order by HireDate desc;
```

```python
%%sql
select  Title, HireDate,Salary,
last_value(FirstName) over(partition by Title order by HireDate rows between unbounded preceding and unbounded following) as Last_Hired
from employees
order by HireDate desc;
```
