```python
%%sql
CREATE TABLE Water_consumption_data_sa (
    City VARCHAR(100),
    Water_consumption INTEGER,
    Date_recorded DATE,
    Time_recorded TIME,
    Timestamp_recorded DATETIME
);
```

```python
%%sql
SELECT * FROM Water_consumption_data_sa;
```

```python
%%sql
select
    City,
    Water_consumption,
    DAY(Date_recorded) AS Day,
    MONTH(Date_recorded) AS Month,
    YEAR(Date_recorded) AS Year
from Water_consumption_data_sa;
```

```python
%%sql
select
    City,
    Date_recorded,
    DATEDIFF(NOW(), Timestamp_recorded) AS Days_since_recorded
from Water_consumption_data_sa;
```

```python
%%sql
SELECT
    City,
    Date_recorded,
    TIMESTAMPDIFF(MONTH, Timestamp_recorded, NOW()) AS Months_since_recorded,
    TIMESTAMPDIFF(YEAR, Timestamp_recorded, NOW()) AS Years_since_recorded
FROM Water_consumption_data_sa;
```

```python
%%sql
select
    City,
    Water_consumption,
    Date_recorded,
    DATE_ADD(Date_recorded, INTERVAL 9 YEAR) AS Date_plus_9_months,
    DATE_SUB(Date_recorded, INTERVAL 7 DAY) AS Date_minus_7_days
from Water_consumption_data_sa;
```
