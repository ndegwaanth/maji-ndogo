```python
%%sql
CREATE TABLE Water_sources_sa_2022 (
    Source_id INT PRIMARY KEY,
    Source_name VARCHAR(100) NOT NULL,
    Water_type VARCHAR(50) NOT NULL,
    Availability VARCHAR(20) NOT NULL
);
```

```python
%%sql
-- UPPERCASE the source names
select
    Source_id,
    UPPER(Source_name) as Source_name,
    Water_type,
    Availability
from water_sources_sa_2022;
```

```python
%%sql
-- lowerCASE the source names
select
    Source_id,
    LOWER(Source_name) as Source_name,
    Water_type,
    Availability
from water_sources_sa_2022;
```

```python
%%sql
-- LTRIM and RTRIM the source names
select
    LTRIM(RTRIM(Water_type)) as NEW_Water_type,
    LTRIM(RTRIM(source_name)) as NEW_Source_name,
    Availability
from water_sources_sa_2022;
```

```python
%%sql
-- LTRIM and RTRIM the source names
select
    LTRIM(Water_type) as NEW_Water_type
from water_sources_sa_2022;
```

```python
%%sql
-- Get the length of the source names
SELECT
    Source_name,
    LENGTH(Source_name) as Name_length
FROM water_sources_sa_2022;
```
