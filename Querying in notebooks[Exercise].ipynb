## Exercise

Run the necessary queries that will provide us with the following information. Compare your queries with the solutions at the end of this notebook.

### Exercise 1

Find the names of all customers from Germany in the Northwind database.

```python
%%sql
PRAGMA table_info(CUSTOMERS);
```

```python
%%sql
SELECT *
FROM Customers
LIMIT 5;
```

```python
%%sql
select ContactName, City, Country
from customers
where Country = 'Germany';
```

### Exercise 2


Find all products in the Northwind database that have a unit price greater than 30 and are discontinued. 

Note: The product has been discontinued when the column `Discontinued` is equal to  1.

```python
%%sql
pragma table_info(Products);
```

```python
%%sql
SELECT *
from Products
limit 5;
```

```python
%%sql
SELECT *
from Products
WHERE UnitPrice > 30 and Discontinued = 1;
```

### Exercise 3


Find the contact names and phone numbers of customers from either 'USA' or 'Canada'.

```python
%%sql
pragma table_info(customers);
```

```python
%%sql
select ContactName, Phone
from customers
where Country = 'USA' or Country = 'Canada';
```

### Exercise 4

From the 'Orders' table, select all orders that were placed by 'VINET', 'QUICK' or 'SUPRD' and were shipped via shipper 1 or 2.


```python
%%sql
pragma table_info(Orders);
```

```python
%%sql
select * from Orders
limit 5;
```

```python
%%sql
SELECT *
from orders
where CustomerID in('VINET', 'QUICK', 'SUPRD') AND (ShipVia = 1 OR ShipVia = 2);
```

### Challenge question

Find the employees in the Northwind database who are either Sales Representatives or Sales Managers and are based in London.


```python
%%sql
pragma table_info(employees);
```

```python
%%sql
SELECT NULLIF ‘hello’,0 ;
```

```python
%%sql
SELECT *
from employees
limit 1;
```

```python
%%sql
select FirstName, LastName, Title, Country
from employees
where (Title = 'Sales Representative' OR Title = 'Sales Manager') AND Country = 'London';
```

## Solutions

### Exercise 1

```python
%%sql
SELECT 
    ContactName 
FROM 
    Customers 
WHERE 
    Country = 'Germany';
```

### Exercise 2

```python
%%sql

SELECT 
    ProductName, Discontinued
FROM 
    Products 
WHERE 
    UnitPrice > 30 
    AND Discontinued = 1;
```

### Exercise 3

```python
%%sql

SELECT 
    ContactName, 
    Phone 
FROM 
    Customers 
WHERE 
    Country IN ('USA', 'Canada');
```

### Exercise 4

```python
%%sql

SELECT 
    * 
FROM 
    Orders 
WHERE 
    CustomerID IN ('VINET', 'QUICK', 'SUPRD')
    AND ShipVia IN (1,2);
```

### Challenge question

```python
%%sql

SELECT 
    FirstName, 
    LastName 
FROM 
    Employees 
WHERE 
    (Title = 'Sales Representative' OR Title = 'Sales Manager') 
    AND City = 'London';
```
